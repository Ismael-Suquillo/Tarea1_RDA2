Escribir un programa que pida al usuario un número entero y muestre por pantalla si es un número primo o no.

```python
aux = 2
n = int(input("Ingrese un número entero positivo:"))
if n > 0:
    while n%aux != 0:
        aux = aux + 1
    if aux == n:
        print(f'El número {n} es primo.')
    else:
        print(f'El número {n} no es primo')
else:
    print("El número debe ser positivo.")
```

Escribir un programa que almacene las matrices en una lista y muestre por pantalla su producto.



Nota: Para representar matrices mediante listas usar listas anidadas, representando cada vector fila en una lista.

```python
A = [[1,2,3],[4,5,6]]
B = [[-1,0],[0,1],[1,1]]
resultado = [[0,0],[0,0]]
for i in range(len(A)):
    for j in range(len(B[0])):
        for k in range (len(B)):
            resultado[i][j] += A[i][k] * B[k][j]
for i in range(len(resultado)):
    resultado[i] = list(resultado[i])
resultado = list(resultado)
for i in range(len(resultado)):
    print(resultado[i])
```

Escribir un programa que pregunte por una muestra de números, separados por comas, los guarde en una lista y muestre por pantalla su media y desviación típica.

```python
muestra = str(input("Ingrese una muestra de números separados por números:"))
muestra = muestra.split(",")
n = len(muestra)
for i in range(n):
    muestra[i] = int(muestra[i])
muestra = tuple(muestra)
suma = 0
raiz_suma = 0
for i in muestra:
    suma += i
    raiz_suma += i**2
media = suma/n
desviacion_st = (raiz_suma/n-media**2)**(1/2)
print(f'La media es {media}, y la desviación típica es: {desviacion_st}')
```
